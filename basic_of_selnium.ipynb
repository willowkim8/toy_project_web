# import package

```python
from selenium import webdriver
import os
```

# 브라우저 driver 설치

Chrome : https://sites.google.com/a/chromium.org/chromedriver/downloads

Selenium 기능 중에서 컴퓨터가 직접 웹브라우저를 띄운 후 코드를 쳐서 동작시킬 수 있도록 'webdriver'라는 api제공한다.  
컴퓨터가 webdriver api로 브라우저를 직접 제어할 수 있도록 도와주는 driver를 설치해주어야 한다.  
각자의 브라우저와 운영체제에 맞는 버전의 드라이버를 다운받으면 된다.

```python
driverPath = os.getenv("HOME")+'/aiffel/toy_project/chromedriver'
driver = webdriver.Chrome(driverPath)
```

  
이런 빈 브라우저 창이 뜬다.

```python
# webdriver가 google 페이지에 접속하도록 명령
driver.get('https://www.google.com')
```

자동화된 테스트 소프트웨어에 의해 제어되고 있음을 알리면서 구글 열었다.

```python
# webdriver를 종료하여 창이 사라진다
driver.close()
```
