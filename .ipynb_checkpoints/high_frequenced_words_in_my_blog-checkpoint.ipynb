# 게시물 URL 크롤링으로 가져오기

참고 : [셀레니움(selenium)으로 크롤링(스크래핑)](https://luminitworld.tistory.com/90)

```python
driverPath = os.getenv("HOME")+"/aiffel/toy_project/chromedriver"
driver = webdriver.Chrome(driverPath)
driver.get("https://blog.naver.com/kin7605")

element = driver.find_element_by_id("mainFrame") # iframe 태그 엘리먼트 찾기
driver.switch_to.frame(element) # 프레임 이동
```

```python
url_set = []
```


xpath 카피하면 안되길래 Full xpath 카피함->그런데 url을 가져온게 아니고 내용만 가져온거 같아서 다른 방법 찾기

```python
# # <strong class="page pcol3 _goPageTop _param(1) => //*[@id="toplistWrapper"]/div[2]/div/strong[2]
# //*[@id="listTopForm"]/table/tbody/tr[1]/td[1]/div/span/a
```

```python
element = driver.find_element_by_xpath("//*[@id='toplistWrapper']/div[2]/div/strong[2]")
driver.execute_script("arguments[0].click();", element)
```

```python
# # orginal 
# # -> driver.find_element_by_xpath("//a[@class='page pcol2 _goPageTop _param(24)']").click()
driver.find_element_by_xpath("//*[@id='toplistWrapper']/div[2]/div/a[1]").click()
```

```python
# link = driver.find_element_by_xpath("//a[@class='pcole _setTop _setTopListUrl']")
link = driver.find_element_by_xpath("//*[@id='listTopForm']/table/tbody/tr[1]/td[1]/div/span/a")
for i in range(len(link)):
    temp = link[i].get_attribute('href')
    temp_ip = temp.split('&')[1].split('=')[-1]
    address = 'https://blog.naver.com/kin7605/' + temp_ip
    url_set.append(address)
    
len(set(url_set))
```

```python
# orginal 
# -> driver.find_element_by_xpath("//a[@class='page pcol2 _goPageTop _param(24)']").click()
driver.find_element_by_xpath("//a[@class='page pcol3 _goPageTop _param(1)']").click()
```
